```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
path="healthcare_dataset.csv"
df=pd.read_csv(path)
df.head()
df
```

```python
disease_a = df['Medical Condition'][:44400].to_numpy()
med_a=df['Medication'][:44400].to_numpy()
gender_a= df['Gender'][:44400].to_numpy()
blood_a= df['Blood Type'][:44400].to_numpy()
age_a= df['Age'][:44400].to_numpy()
```

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

text_data = np.column_stack([med_a, gender_a, blood_a])

# One-hot encode the text data
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
text_encoded = encoder.fit_transform(text_data)
y_train = encoder.fit_transform(disease_a.reshape(-1, 1))
print("Shape of encoded text data:", text_encoded.shape)
print("Shape of encoded text data:", y_train.shape)
```

```python
from sklearn.preprocessing import StandardScaler

# Reshape age_a to a 2D array and scale
scaler = StandardScaler()
age_scaled = scaler.fit_transform(age_a.reshape(-1, 1))
```

```python
X_train = np.hstack([text_encoded, age_scaled])
print("Shape of combined feature matrix:", X_train.shape)
```

```python
disease_b = df['Medical Condition'][44400:55500].to_numpy()
med_b=df['Medication'][44400:55500].to_numpy()
gender_b= df['Gender'][44400:55500].to_numpy()
blood_b= df['Blood Type'][44400:55500].to_numpy()
age_b= df['Age'][44400:55500].to_numpy()
```

```python
text_data_2 = np.column_stack([med_b, gender_b, blood_b])

text_encoded_2 = encoder.fit_transform(text_data_2)
y_test = encoder.fit_transform(disease_b.reshape(-1, 1))
print("Shape of encoded text data:", text_encoded_2.shape)
print("Shape of encoded text data:", y_test.shape)
```

```python
age_scaled_2 = scaler.fit_transform(age_b.reshape(-1, 1))
X_test = np.hstack([text_encoded_2, age_scaled_2])
print("Shape of combined feature matrix:", X_test.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input

model = Sequential([
    Input(shape=(16,)),
    Dense(units=400, activation='relu'),
    Dense(units=200, activation='relu'),
    Dense(units=50, activation='relu'),
    Dense(units=6, activation='softmax')  
])
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=1500, validation_split=0.1)
```

```python
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test accuracy: {test_accuracy:.4f}")
```

```python
model.save("disease_model_2.keras")
```
